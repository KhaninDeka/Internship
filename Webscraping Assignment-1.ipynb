```python
# Q1: Write a python program to display all the header tags from wikipedia.org.

def Wiki(url):
    r = requests.get(url)
    soup = bs(r.content)       
    headers_wiki = soup.find_all(["h1","h2","h3","h4","h5","h6"])
    
    for i in headers_wiki:
        print(i,"\n")

Headers_wiki= Wiki("https://en.wikipedia.org/wiki/Main_Page")
```

```python
# Q2: Write a python program to display IMDBâs Top rated 100 moviesâ data (i.e. name, rating, year of release) and make a dataframe

def imdb(url):
    r = requests.get(url)
    soup = bs(r.content)
    
    titles = soup.find_all("td",class_="titleColumn",limit=100) 
    movie_names=[]
    for x in titles:
        name = x.find('a').text
        movie_names.append(name)
        
    movie_years= soup.find_all("span", class_="secondaryInfo",limit=100)
    years=[]
    for x in movie_years:
        year = x.get_text()[1:5]
        years.append(year)
    
    movie_ratings= soup.find_all("strong",limit=100)
    ratings= []
    for x in movie_ratings:
        rating= x.get_text()
        ratings.append(rating)
        
    imdb_movies= pd.DataFrame({'Title':movie_names,'Rating':ratings,'Year':years})
    
    return imdb_movies

imdb_100 = imdb("https://www.imdb.com/chart/top/?ref_=nv_mv_250")
imdb_100.style.hide_index()
```

```python
# Q4: Write s python program to display list of respected former presidents of India(i.e. Name , Term of office) from https://presidentofindia.nic.in/former-presidents.htm

def Presidents_India(url):
    r = requests.get(url)
    soup = bs(r.content)
    
    presidents= soup.find_all("div",class_="presidentListing")

    names= []
    for x in presidents:
        name_of_prez=x.find("h3")
        name = name_of_prez.get_text()
        name_= name.split("(")
        names.append(name_[0])
        
        
    terms= []
    for x in presidents:
        president= x.find("p").get_text()[16:]
        terms.append(president)
        
    presidents= pd.DataFrame({'Name':names,'Term':terms})
    
    return presidents

Presidents = Presidents_India("https://presidentofindia.nic.in/former-presidents.htm")
Presidents.style.hide_index()
```

```python
# Q5(a): Top 10 ODI teams in menâs cricket along with the records for matches, points and rating.

def ICC_Teams(url):
    r = requests.get(url)
    soup = bs(r.content)
    
    teams= soup.find_all("span",class_="u-hide-phablet", limit=10)
    team_names=[]
    for x in teams:
        team= x.get_text()
        team_names.append(team)
             
        
    matches_rank1= soup.find("td",class_="rankings-block__banner--matches").get_text()
    points_rank1= soup.find("td",class_="rankings-block__banner--points").get_text()
    rating_rank1= soup.find("td",class_="rankings-block__banner--rating u-text-right").get_text()[29:32]
    rank1_data= [team_names[0],matches_rank1,points_rank1,rating_rank1]
    team_names.pop(0)
        
    matches_points= soup.find_all("td",class_="table-body__cell u-center-text",limit=18)
    combined_list=[]

    for i in matches_points:
        mp= i.get_text()
        combined_list.append(mp)
    
    matches= combined_list[::2]
    points= combined_list[1::2]

    ratings_data= soup.find_all("td",class_="table-body__cell u-text-right rating",limit=9)
    ratings_list=[]
    for i in ratings_data:
        rating= i.get_text()
        ratings_list.append(rating)  
        

    teams_rank2_onwards= pd.DataFrame({'Country':team_names,'Matches':matches,'Points':points,'Rating':ratings_list})
        
    df_rank1= pd.DataFrame({'Country':[rank1_data[0]],'Matches':[rank1_data[1]],'Points':[rank1_data[2]],'Rating':[rank1_data[3]]})   
        
    final_df= pd.concat([df_rank1,teams_rank2_onwards],axis=0)  
               
    
    return final_df

ICC_Top_Teams = ICC_Teams("https://www.icc-cricket.com/rankings/mens/team-rankings/odi")
ICC_Top_Teams.style.hide_index()
```

```python
# Q5(c): Top 10 ODI bowlers along with the records of their team and rating.

def ICC_Bowler(url):
    r = requests.get(url)
    soup = bs(r.content)
    
    rank1_name= soup.find_all("div",class_="rankings-block__banner--name")
    rank1_name= rank1_name[1].get_text()
    rank1_country= soup.find_all("div",class_="rankings-block__banner--nationality")
    rank1_country= rank1_country[1].get_text()[2:5]
    rank1_rating= soup.find_all("div",class_="rankings-block__banner--nationality")
    rank1_rating= rank1_rating[1].get_text()[33:36]

    rank1_df= pd.DataFrame({'Name':[rank1_name],'Team':[rank1_country],'Rating':[rank1_rating]})
    
    names= soup.find_all("td",class_="table-body__cell name",limit=18)
    player_names=[]
    count=0
    for i in names:
        if (count>=9):
            name= i.get_text()
            name= name.split("\n")
            player_names.append(name[1])
        count+=1
        
    countries= soup.find_all("span",class_="table-body__logo-text",limit=18)
    player_countries=[]
    count=0
    for i in countries:
        if (count>=9):
            country= i.get_text()
            player_countries.append(country)
        count+=1
        
    ratings= soup.find_all("td",class_="table-body__cell u-text-right rating",limit=18)
    player_ratings=[]
    count=0
    for i in ratings:
        if count>=9:
            rating= i.get_text()
            player_ratings.append(rating)
        count+=1
      
    rank2_onwards_df= pd.DataFrame({'Name':player_names,'Team':player_countries,'Rating':player_ratings})
    
    final_df= pd.concat([rank1_df,rank2_onwards_df],axis=0)
    
    return(final_df)


ICC_Top_Bowler = ICC_Bowler("https://www.icc-cricket.com/rankings/mens/player-rankings/odi")
ICC_Top_Bowler.style.hide_index() 
```

```python
# Q7: Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world : 
# i) Headline ii) Time iii) News Link

def CNBC_Headlines(url):
    
    r = requests.get(url)
    soup = bs(r.content)
    
    headlines= soup.find_all("div",class_="LatestNews-container")
    
    for i in headlines:
        headline= i.find("a",class_="LatestNews-headline").get_text()
        print(headline)
        time= i.find("time",class_="LatestNews-timestamp").get_text()
        print("Published",time)
        link= i.find("a",class_="LatestNews-headline", href= True)
        print("Link:",link['href'],"\n")
        
Headlines= CNBC_Headlines("https://www.cnbc.com/world/?region=world")
    
```

```python
# Q8: Write a python program to scrape the details of most downloaded articles from AI in last 90 days. https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles
# Scrape below mentioned details :
# i) Paper Title
# ii) Authors
# iii) Published Date
# iv) Paper URL

def Journals(url):
    r= requests.get(url)
    soup= bs(r.content)

    data= soup.find_all("li",class_="sc-9zxyh7-1 sc-9zxyh7-2 exAXfr jQmQZp")

    for i in data:

        paper_title= i.find("h2",class_="sc-1qrq3sd-1 MKjKb sc-1nmom32-0 sc-1nmom32-1 hqhUYH ebTA-dR").get_text()
        print("Title:",paper_title)

        author_data= i.find("span",class_="sc-1w3fpd7-0 pgLAT").get_text()
        print("Author(s):",author_data)

        date_data= i.find("span",class_="sc-1thf9ly-2 bKddwo").get_text()
        print("Published:",date_data)

        link= i.find("a",class_="sc-5smygv-0 nrDZj", href= True)
        print("Link:",link["href"],"\n")
        
AI_Journals= Journals("https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles")
```

```python
# Q9: Write a python program to scrape mentioned details from dineout.co.in :
# i) Restaurant name
# ii) Cuisine
# iii) Location
# iv) Ratings
# v) Image URL

def Restaurants(url):

    r= requests.get(url)
    soup= bs(r.content)

    data= soup.find_all("div",class_="restnt-main-wrap clearfix")
    
    for i in data:

        name= i.find("a",class_="restnt-name ellipsis").get_text()
        print(" ",name)

        cuisine_data= i.find("span",class_="double-line-ellipsis").get_text()
        cuisine= cuisine_data.split("|")
        print("",cuisine[1])

        location= i.find("div",class_="restnt-loc ellipsis").get_text()
        print(" ",location)

        rating= i.find("div",class_="restnt-rating rating-4").get_text()
        print(" ",rating)

        restaurant_picture= i.find("img",class_="no-img")
        print("  Picture Link: \n",restaurant_picture['data-src'],"\n")
    
Restaurant_Details= Restaurants("https://www.dineout.co.in/delhi-restaurants/buffet-special")
```

```python
# Q10: Write a python program to scrape the details of top publications from Google Scholar from https://scholar.google.com/citations?view_op=top_venues&hl=en
# i) Rank
# ii) Publication
# iii) h5-index
# iv) h5-median

def google_scholar(url):
    
    r= requests.get(url)
    soup= bs(r.content)
    
    rank_data= soup.find_all("td",class_="gsc_mvt_p")
    ranks=[]

    for i in rank_data:
        rank= i.get_text()
        ranks.append(rank)
        
    pub_data= soup.find_all("td",class_="gsc_mvt_t")
    pubs=[]

    for i in pub_data:
        pub= i.get_text()
        pubs.append(pub)
    
    index_median_data= soup.find_all("td",class_="gsc_mvt_n")
    combined_data=[]

    for i in index_median_data:
        im= i.get_text()
        combined_data.append(im)

    h5_index= combined_data[::2]
    h5_median= combined_data[1::2]
    
    final_df= pd.DataFrame({'Rank':ranks,'Publication':pubs,'h5-index':h5_index,'h5-median':h5_median})
    
    return(final_df)

Google_Scholar_Pubs= google_scholar("https://scholar.google.com/citations?view_op=top_venues&hl=en")
Google_Scholar_Pubs.style.hide_index()
```
