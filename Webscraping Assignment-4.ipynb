```python
# Q1 Solution:

# defining the lists to be used.
ranks=[]
names=[]
artists=[]
upload_date=[]
views=[]

# defining the function for getting the target data.
def YT_Most_Viewed(url):
    
    # opening the target website.
    driver.get(url)
    time.sleep(1)
    
    # using the explicit wait method till the data-table is loaded.
    wait= WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter']")))
    # finding the webelement for the table and getting the text data.
    data = driver.find_element(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter']").text
    
    components= data.split("\n")    # splitting the text data based on newline character.
    count=1
    
    for i in components[2:]:
        if i==components[-1]:
            pass
        else:
            ranks.append(i.split('"')[0].strip()) 
            names.append(i.split('"')[1].strip())
            
            # getting the data of artists using X-Path and try-except blocks.
            try:
                artist_data=driver.find_element(By.XPATH,f'//*[@id="mw-content-text"]/div[1]/table[2]/tbody/tr[{count}]/td[3]/a')
                artists.append(artist_data.text)
            except NoSuchElementException:
                artist_data=driver.find_element(By.XPATH,f'//*[@id="mw-content-text"]/div[1]/table[2]/tbody/tr[{count}]/td[3]')
                artists.append(artist_data.text)
                
            # getting the views data using X-Path.
            views_data=driver.find_element(By.XPATH,f'//*[@id="mw-content-text"]/div[1]/table[2]/tbody/tr[{count}]/td[4]')
            views.append(views_data.text)

            # getting the upload-dates using X-Path.
            date_data=driver.find_element(By.XPATH,f'//*[@id="mw-content-text"]/div[1]/table[2]/tbody/tr[{count}]/td[5]')
            upload_date.append(date_data.text)
            count+=1

    # the final dataframe.
    results= pd.DataFrame({"Rank":ranks,
                            "Name":names, 
                            "Artist/Uploader":artists, 
                            "Upload Date":upload_date, 
                            "Views(billions)":views
                          })
    return results


driver= webdriver.Chrome("chromedriver.exe")     # connecting to the webdriver.
url="https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"      # mentioning the url to scrape data from. 
df= YT_Most_Viewed(url)    # calling the function and saving the final dataframe.

driver.close()    # closing the webdriver.
df
```

```python
# Q2 Solution:

# defining the lists to be used.
titles=[]
series=[]
venues=[]
date=[]
time_=[]

# defining the function for getting the target data.
def TeamIndia_Fixtures(url):
    
    r= requests.get(url)     # sending request to the target website.
    soup= bs(r.content)      # getting the parsed content from which we will extract our data. 
    
    # getting the link for the page containing the fixtures data.
    data= soup.find("div",id="navigation")
    link= data.find("a",class_="nav-link",href=True)

    driver.get(link['href'])    # connecting to the target page.
    time.sleep(2)

    # using the explicit wait method till the fixture cards are loaded.
    wait= WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,"//div[@class='fixture-card-top']")))
    # finding the webelements with fixtures data.
    data= driver.find_elements(By.XPATH,"//div[@class='fixture-card-top']")    
    for i in data:
        components=i.text.split("\n")       # processing and appending the relevant data into respective lists.
        series.append(components[0])
        date.append(components[1])
        time_.append(components[2])
    
    # finding the webelements with fixtures data. 
    data= driver.find_elements(By.XPATH,"//div[@class='fixture-card-bottom']")
    for i in data:
        components=i.text.split("\n")                       # processing and appending the relevant data into respective lists.
        titles.append(components[0].split("-")[0].strip())
        venues.append(components[0].split("-")[1].strip())

    # the final dataframe.
    results= pd.DataFrame({"Match Title":titles,
                           "Series":series, 
                           "Place":venues, 
                           "Date":date, 
                           "Time":time_
                           })
    return results

url="https://www.bcci.tv/"                       # mentioning the url to scrap data from.
driver= webdriver.Chrome("chromedriver.exe")     # connecting to the webdriver.
df= TeamIndia_Fixtures(url)                      # calling the function and saving the final dataframe.

driver.close()     # closing the webdriver.
df
```

```python
# Q3 Solution:

# defining a 'User-Agent' and 'Accept-Language' for our requests.
HEADERS = ({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:102.0) Gecko/20100101 Firefox/102.0',
            'Accept-Language': 'en-US, en;q=0.5'})

# defining the lists to be used.
names=[]
descriptions=[]

# defining the function for getting the target data.
def guru99_Selenium_Exceptions(url):
    
    r= requests.get(url,headers=HEADERS)     # sending request to the target website.
    soup=bs(r.content)                       # getting the parsed content from which we will extract our data.

    link_data=soup.find("a",title="Selenium",href=True)       # getting the link-data to "Selenium" page using tag and title.

    r= requests.get(url+link_data['href'], headers=HEADERS)   # sending request to the "Selenium" page.
    soup= bs(r.content)                                       # getting the parsed content
    
    # getting the link-data to "Selenium Exception Handling" page using tag and title.
    link_data= soup.find("a",title="Selenium Exception Handling (Common Exceptions List)",href=True)

    r= requests.get(url+link_data['href'], headers=HEADERS)   # sending request to the "Selenium Exception Handling" page.
    soup= bs(r.content)                                       # getting the parsed content
    
    # finding the webelement contaning our target table of "Selenium Exception" and getting the text data.
    data= soup.find("table",class_="table table-striped").text  
    contents=data.strip().split("\n\n\n")                           
    for i in contents[1:]:
        elements= i.split("\n")                   # processing and extracting the relevant data into respective lists.
        names.append(elements[0])
        descriptions.append(elements[1])

    # the final dataframe.
    results= pd.DataFrame({"Name":names,
                           "Description":descriptions
                          })
    return results

url= "https://www.guru99.com"           # mentioning the url to scrap data from.
df= guru99_Selenium_Exceptions(url)     # calling the function and saving the final dataframe.
df
```

```python
# Q5 Solution:

# defining the lists to be used.
titles=[]
descriptions=[]
contributors=[]
languages=[]

# defining the function for getting the target data.
def Git_Repos(url):
    
    driver.get(url)     # connecting to the target website.

    # using the explicit wait method till the "Explore" button gets loaded.
    wait= WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,"/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details")))
    # clicking on the "Explore" button.
    explore_btn= driver.find_element(By.XPATH,"/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details").click()

    # using the explicit wait method till the "Trending" button gets loaded.
    wait= WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,"/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul/li[5]/a")))
    # getting the webelement for the "Trending" button.
    trending_btn= driver.find_element(By.XPATH,"/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul/li[5]/a")
    link= trending_btn.get_attribute('href')   # saving the link to the "Trending" page.
    
    r= requests.get(link)         # sending request to the "Trending" page.
    soup=bs(r.content)            # getting the parsed content from which we will extract our data.               

    data= soup.find_all("article",class_="Box-row")     # finding the webelements for the different repositories.
    
    # running a for loop over each repository webelement.
    for i in data:
        repo_title_data= i.find("h1")                      
        link_data= repo_title_data.find("a",href=True)            # getting the link to the page of that repository.

        r= requests.get("https://github.com"+link_data['href'])   # sending the request to the page.
        soup_new= bs(r.content)                                   # getting the parsed content.

        # finding the title data and extracting it to a list.
        title_data= soup_new.find("strong",class_="mr-2 flex-self-stretch")    
        titles.append(title_data.text.strip())

        # getting the description data and extracting it to a list.
        description_data=soup_new.find("p",class_="f4 my-3")       
        try:
            descriptions.append(description_data.text.strip())
        except AttributeError:
            descriptions.append("-")         # in case of missing data "-" is appended.                            

        # getting the no. of contributors data and extracting it to a list.
        contributors_data= soup_new.find_all("h2",class_="h4 mb-3")     
        for i in contributors_data:
            element=i.text.strip()
            check= element.find("Contributors")
            if check!=(-1):
                contributors.append(element.split("Contributors")[-1].strip())
                break
            elif i==contributors_data[-1]:
                contributors.append("-")
            else:
                continue
        
        # getting the languages-used data and extracting it to a list.
        languages_data= soup_new.find_all("div",class_="BorderGrid-cell")   
        for i in languages_data:
            element=i.text.strip()
            check= element.find("Languages")
            if check!=(-1):
                languages_list=[]
                x= element.split("Languages")[-1].strip().split("\n\n\n\n\n\n\n\n")
                for j in x:
                    languages_list.append(j.split("\n")[0])
                languages_string=",".join(languages_list)
                languages.append(languages_string)
                break
            elif i==languages_data[-1]:
                languages.append("-")
            else:
                continue

    # the final dataframe
    results= pd.DataFrame({"Title":titles,
                           "Description":descriptions, 
                           "Contributors Count":contributors, 
                           "Language Used":languages
                           })
    return results

url="https://github.com"         # mentioning the url to scrap data from.
driver= webdriver.Chrome("chromedriver.exe")       # connecting to the webdriver.
df= Git_Repos(url)          # calling the function and saving the final dataframe.

driver.close()      # closing the webdriver.
df
```

```python
# Q6 Solution:

# defining the lists to be used.
name=[]
artist=[]
lw_rank=[]
pk_rank=[]
wks_onbd=[]

# defining the function for getting the target data.
def Billboard_100(url):
    
    try:
        driver.get(url)         # connecting to the target website.
    except WebDriverException:
        pass

    time.sleep(2)

    try:
        # using the explicit wait method till the "Charts" button gets loaded.
        wait= WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,"/html/body/div[3]/header/div[1]/div/div/div[2]/div/nav/ul/li[1]/a")))
        # clicking on the "Charts" button.
        charts_btn= driver.find_element(By.XPATH,"/html/body/div[3]/header/div[1]/div/div/div[2]/div/nav/ul/li[1]/a").click()
    except WebDriverException:
        pass

    time.sleep(2)

    try:
        # using the explicit wait method till the "Hot 100" button gets loaded.
        wait= WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,"/html/body/div[3]/header/div[2]/div/nav/ul/li[1]/a")))
        # finding the webelement for the "Hot 100" button.
        hot_100_btn= driver.find_element(By.XPATH,"/html/body/div[3]/header/div[2]/div/nav/ul/li[1]/a")
        link= hot_100_btn.get_attribute("href")   # getting the link to the "Hot 100" page.
        hot_100_btn.click()    # clicking on the "Hot 100" button.
    except WebDriverException:
        pass

    time.sleep(2)

    try:
        r= requests.get(driver.current_url)      # sending request to the target page.
    except WebDriverException:
        pass

    soup= bs(r.content)        # getting the parsed content.
    
    # finding the webelements containing the songs data.
    data= soup.find_all("div",class_="o-chart-results-list-row-container")  

    # processing each song webelement and extracting the relevant data into respective lists. 
    for i in data:
        song_data= i.find("li",class_="lrv-u-width-100p").text.strip().replace("\n","").replace("\t\t\t\t\t","/").replace("\t\t","/").split("/")
        name.append(song_data[0])
        artist.append(song_data[1])
        wks_onbd.append(song_data.pop())
        pk_rank.append(song_data.pop())
        lw_rank.append(song_data.pop())

    # the final dataframe.
    results= pd.DataFrame({"Song Name":name,
                           "Artist":artist, 
                           "Last Week Rank":lw_rank, 
                           "Peak Rank":pk_rank,
                           "Weeks on chart":wks_onbd
                           })
    
    return results


url="https://www.billboard.com"                  # mentioning the url to scrap data from.
driver= webdriver.Chrome("chromedriver.exe")     # connecting to the webdriver.
df= Billboard_100(url)            # calling the function and saving the final dataframe.

driver.close()        # closing the webdriver.
df
```

```python
# Q8 Solution:

# defining the lists to be used.
name=[]
author=[]
vol_sold=[]
pub=[]
genre=[]

# defining the function for getting the target data.
def Books(url):
    
    r= requests.get(url)      # sending request to the target page.
    soup= bs(r.content)       # getting the parsed content.

    table_data= soup.find("table",class_="in-article sortable")   # getting the webelement for "Data summary" table.
    data= table_data.find_all("tr")           # getting the webelements containing the data about books.
    
    # processing each book webelement and extracting the relevant data into respective lists. 
    for i in data[2:]:
        elements= i.text.strip().split(" \n ")
        genre.append(elements.pop())
        pub.append(elements.pop())
        vol_sold.append(elements.pop())
        author.append(elements.pop())
        name.append(elements.pop())

    # the final dataframe
    results= pd.DataFrame({"Book Name":name,
                           "Author":author, 
                           "Volume Sales":vol_sold, 
                           "Publisher":pub,
                           "Genre":genre
                           })
    return results

# mentioning the target url.
url="https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare"
df= Books(url)    # calling the function and saving the final dataframe.
df
```

```python
# Q9 Solution:

# defining the lists to be used.
name=[]
yr_span=[]
genre=[]
runtime=[]
rating=[]
votes=[]

# defining the function for getting the target data.
def IMDB_Series(url):
    
    r= requests.get(url)       # sending request to the target page.
    soup= bs(r.content)        # getting the parsed content.

    # finding the webelements containing the data about tv-series.
    data= soup.find_all("div",class_='lister-item-content')
    # processing each tv-series webelement and extracting the relevant data into respective lists.
    for i in data:
        elements= i.text.strip().split("|")
        name.append(elements[0].strip().split("\n")[1])
        yr_span.append(elements[0].strip().split("\n")[2])    
        runtime.append(i.find("span",class_="runtime").text.strip())
        genre.append(i.find("span",class_="genre").text.strip())
        votes.append(i.text.strip().split("\n")[-1])
        rating.append(i.find("span",class_="ipl-rating-star__rating").text)

    # the final dataframe
    results= pd.DataFrame({"Name":name,
                           "Year span":yr_span, 
                           "Genre":genre, 
                           "Runtime":runtime,
                           "Rating":rating,
                           "Votes":votes
                           })
    return results

url="https://www.imdb.com/list/ls095964455/"       # mentioning the url to scrap data from.  
df= IMDB_Series(url)           # calling the function and saving the final dataframe.
df
```

```python
# Q10 Solution:

# defining the lists to be used.
names=[]
type_=[]
task=[]
attr_type=[]
no_inst=[]
no_attr=[]
year=[]

# defining the function for getting the target data.
def UCI_Databases(url):
    
    driver.get(url)      # connecting to the target website.
    time.sleep(1)

    # using the explicit wait method till the "View ALL Data Sets" button is loaded.
    wait= WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,"/html/body/table[1]/tbody/tr/td[2]/span[2]/a")))
    # clicking on the "View ALL Data Sets" button.
    driver.find_element(By.XPATH,"/html/body/table[1]/tbody/tr/td[2]/span[2]/a").click()

    time.sleep(3)

    r= requests.get(driver.current_url)     # sending request to the webdriver's current page.  
    soup= bs(r.content,"html.parser")       # getting the parsed content.

    # finding the webelements for the different tables in the page.
    tables_data=soup.find_all("table")
    # choosing the webelement having the table of datasets and then finding the webelements containing the info about various datasets. 
    data= tables_data[1].find_all("tr")     
    
    # processing each dataset-info webelement and extracting the relevant data into respective lists.
    for i in data[18::2]:
        elements= i.find_all("p",class_="normal")
        names.append(elements[0].text.strip())
        type_.append(elements[1].text.strip())
        task.append(elements[2].text.strip())
        attr_type.append(elements[3].text.strip())
        no_inst.append(elements[4].text.strip())
        no_attr.append(elements[5].text.strip())
        year.append(elements[6].text.strip())

    # the final dataframe
    results= pd.DataFrame({"Dataset Name":names,
                           "Data Type":type_, 
                           "Task":task, 
                           "Attribute Type":attr_type,
                           "No. of instances":no_inst,
                           "No. of attributes":no_attr,
                           "Year": year
                           })
    return results
    
url="https://archive.ics.uci.edu/"            # mentioning the url to scrap data from.   
driver= webdriver.Chrome("chromedriver.exe")  # connecting to the webdriver.
df= UCI_Databases(url)        # calling the function and saving the final dataframe.

driver.close()          # closing the webdriver.
df
```
